```python
# Import necessary libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# make_blobs is a function provided by scikit learn to create synthetic datasets.
```

```python
X , y = make_blobs(n_samples=500 , centers=6 , cluster_std=0.8 , random_state=23)
```

```python
# n_samples -> no. of samples
# centers -> no. of centers you want in your data
# cluster_std -> std deviation of clusters
```

```python
plt.scatter(X[:,0],X[:,1],s=10)
```

```python
# s -> size of a data point
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(X)
```

```python
# n_clusters -> no. of clusters to make in data
```

```python
y_pred = kmeans.predict(X)
```

```python
y_pred.shape
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred,s=10)
```

```python
# c -> list of colors
```
